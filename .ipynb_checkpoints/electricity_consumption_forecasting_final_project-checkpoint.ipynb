```python
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
warnings.filterwarnings("ignore")

# decompose the series into trend, seasonality, and residual

result = seasonal_decompose(power_df.usage, model='add', period=365)
obs, trend, snl, resid = result.observed, result.trend, result.seasonal, result.resid
```

```python
# plot the seasonal decomposotion graph

def plot_series_comps(obs, trend, seasonal, resid, series_name):

  labels = "Observed", "Trend", "Seasonality", "Residuals"
  comps = obs, trend, seasonal, resid
  colors = "black", "blue", "orange", "red"

  fig, axes = plt.subplots(4, 1, figsize=(18, 12))

  for ax, comp, label, cl in zip(axes.flatten(), comps, labels, colors):
    ax.plot(comp, color=cl)
    ax.set_xlim([trend.index.min(), trend.index.max()])
    title = "%s: %s" % (series_name, label)
    ax.set_title(title, size=16)

  plt.subplots_adjust(hspace=0.5)


plot_series_comps(obs, trend, snl, resid, 
                  series_name="Residential Daily-Average Power Usage")
```

**Insights**

From the seasonal decomposition graph, it's apparent that:

* The series have a clear seasonal component. That is, every year the residential electric consumption follows a concave downward curve where highest consumption is around the months of May to October.

* On average, electric consumption decreased in 2017 and begun to increase again in 2018 and is slightly decreasing in 2019 until beginning of 2020 as implied by the trend component.

## Train-Test-Split and Data Preprocessing

```python
# merge two dataframe with the date index as the join key

df = pd.merge(power_df, weather_df, left_index=True, right_index=True).drop(
    ["day_of_week"], axis=1)
df.head()
```

Insights

For the current dataset, there are 97 covid-lockdowns recorded. To make the testing dataset to be representative as possible of the training dataset, we divide this number into reasonable sizes for training, validation, and testing dataset:

* Training - 50 covid-lockdowns
* Validation - 20 covid-lockdowns
* Testing - 27 covid-lockdowns

```python
# get records on year 2020
rec_2020 = df.loc["2020"]

# fetch the first 80% of data (this would result in 71 lockdowns in training and 26 lockdowns in testing)
size = int(len(rec_2020) * 0.80)
train2020, test2020 = rec_2020[:size], rec_2020[size:]
train2020[train2020.notes == "COVID_lockdown"].shape
```

Insights

As can be seen, there are 71 covid-lockdowns in year 2020 for training/validation and remaining 26 covid-lockdowns in testing.

```python
train2020.index.min(), train2020.index.max()
```

Insights

As can be seen, the 2020 period range for training/validation is 1st January to 30th April. Because of this, we use the latter date as the cut-off date for the final full training and testing split and use the first 98% of full training for training dataset alone and remaining 2% for validation. This would result in a reasonable split of Covid-lockdown days in training, validation, and testing set:

```python
# train-test-split

ftrain, test = df.loc[:"2020-05-30"], df.loc["2020-05-31":]

# train-validation-split

size = int(len(ftrain) * 0.98)
train, val = ftrain.iloc[:size], ftrain.iloc[size:]
```

As can be seen, the number of the COVID lockdown days in training, validation, and testing is the same as the note above. However, it's also apparent that there are no vacation days in both validation and testing datasets which could lead to misrepresentation of training set of both validation and testing set.
```python
# period range for training, validation, and test dataset

tr_prange = train.index.min(), train.index.max()
v_prange = val.index.min(), val.index.max()
ts_prange = test.index.min(), test.index.max()

print("train period range: {}".format(tr_prange))
print("val period range: {}".format(v_prange))
print("test period range: {}".format(ts_prange))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

num_attribs = list(train.select_dtypes(include=np.float64).columns)
cat_attribs = ["notes"]

full_pipeline = ColumnTransformer([
  ("num", StandardScaler(), num_attribs),
  ("cat", OneHotEncoder(), cat_attribs),
])

# transform train, test, and validation set
train_tf = pd.DataFrame(full_pipeline.fit_transform(train), index=train.index)
test_tf = pd.DataFrame(full_pipeline.transform(test), index=test.index)
val_tf = pd.DataFrame(full_pipeline.transform(val), index=val.index)

# update the columns
columns = num_attribs + list(full_pipeline.named_transformers_["cat"].categories_[0])
train_tf.columns = columns
test_tf.columns = columns
val_tf.columns = columns

# fit-transform the full training set also (for ARIMAX and SARIMAX modeling)
ftrain_tf = pd.DataFrame(full_pipeline.fit_transform(ftrain), index=ftrain.index)
ftrain_tf.columns = columns
```

## Stationarity Check: Histogram and Augmented Dickey-Fuller Test

```python
def plot_hist(series, title, xlabel, bins=50):
  
  series.hist(bins=bins, figsize=(8, 6))
  title = "Histogram: %s" % title
  plt.title(title, size=14)
  plt.xlabel(xlabel, size=12)
  plt.ylabel("Frequency", size=12)
  plt.grid(True)

plot_hist(ftrain_tf.usage, title="Residential Daily-Average Electricity Consumption",
          xlabel="Daily-Average Power Usage (kWh)")
```

```python
import statsmodels.tsa.stattools as sts

def adfuller_test(series, series_name):
  
  adf, pvalue, usedlag, nobs, critical_values, icbest = sts.adfuller(series)
  series_name = "Augmented Dickey-Fuller Test: %s" % (series_name)
  print(series_name)
  print("=" * len(series_name))
  print("ADF: ", adf)
  print("P-value: ", pvalue)
  print("Nobs: ", nobs)
  print("Critical Values: ", critical_values)
  print("=" * len(series_name))

adfuller_test(ftrain_tf.usage, 
              series_name="Residential Daily-Average Electricity Consumption")
```

## ARIMAX and SARIMAX Modeling

### ARIMAX Modeling

```python
# apply differencing first to the usage column

usage_diff = ftrain_tf.usage.diff()
usage_diff.head()
```

```python
# apply adfuller test to differenced data

adfuller_test(usage_diff[1:], 
              series_name="Diff. Residential Daily-Average Electricity Consumption")
```

Insights

The differenced data passed the test even at 1% level implying that the data is already stationary.

```python
import statsmodels.graphics.tsaplots as sgt

# check acf and pacf plot to see reasonable lag for AR and MA components

fig, ax = plt.subplots(2, 1, figsize=(8, 12))

title = "Residential Daily-Average Electricity Consumption Diff."

sgt.plot_pacf(usage_diff[1:], zero=False, lags=40, ax=ax[0])
ax[0].set_title("PACF: %s" % title, size=14)

sgt.plot_acf(usage_diff[1:], zero=False, lags=40, ax=ax[1])
ax[1].set_title("ACF:%s" % title, size=14)

plt.subplots_adjust(wspace=0.1)
```

Insights

* It seems that incorporating up to 6 lags in the AR component and 7 lags in the MA component in `auto_arima` may result in a better predictor.

```python
from pmdarima import auto_arima

val_size = int(len(ftrain) * 0.2)

# use auto_arima to automate ARIMAX model selection

arimax_model = auto_arima(
    ftrain_tf.usage, exogenous=ftrain_tf.drop("usage", axis=1), max_order=None, 
    start_p=1, start_q=1, max_p=6, max_q=6, max_d=2, maxiter=100, alpha=0.05,  
    njobs=-1, seasonal=False, trace=True, information_criterion="oob", 
    out_of_sample_size=val_size)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# fitting the optimal ARIMAX(1,1,1) model
arimax_model = ARIMA(ftrain_tf.usage, exog=ftrain_tf.drop("usage", axis=1), 
                     order=(1,1,1)).fit()
print(arimax_model.summary())
```

```python
# obtain predictions on training set

start = 0
end = len(ftrain_tf) - 1

arimax_model_tr_preds = arimax_model.predict(
    start=start, end=end, exog=ftrain_tf.drop("usage", axis=1))

fig, ax = plt.subplots(figsize=(12, 4))
ftrain_tf.usage.plot(label="test", color='blue', ax=ax)
arimax_model_tr_preds.plot(label="ARIMAX(1,1,1) train predictions", color='orange', ax=ax)
ax.legend(loc="best")
```

```python
# obtain predictions

start = len(ftrain_tf)
end = len(ftrain_tf) + len(test_tf) - 1

arimax_model_preds = arimax_model.predict(
    start=start, end=end, exog=test_tf.drop("usage", axis=1), )

fig, ax = plt.subplots(figsize=(12, 4))
test_tf.usage.plot(label="test", color='blue', ax=ax)
arimax_model_preds.plot(label="ARIMAX(1,1,1) Predictions", color='orange', ax=ax)
ax.legend(loc="best")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

arimax_model_mae = mean_absolute_error(test_tf.usage, arimax_model_preds)
arimax_model_mse = mean_squared_error(test_tf.usage, arimax_model_preds)
arimax_model_rmse = np.sqrt(arimax_model_mse)
print("ARIMAX(1,1,1) MAE: %f"  % arimax_model_mae)
print("ARIMAX(1,1,1) MSE: %f"  % arimax_model_mse)
print("ARIMAX(1,1,1) RMSE: %f"  % arimax_model_rmse)
```

## Deep Learning: LSTM Model

### Data Preparation: `TimeSeriesGenerator`

```python
from keras.preprocessing.sequence import TimeseriesGenerator

# setup training, validation, and testing time-series generators
n_input = 3
n_features = train_tf.shape[1]

# train generator
train_gen = TimeseriesGenerator(
    train_tf, train_tf.usage, length=n_input, batch_size=32)

# validation generator
val_gen = TimeseriesGenerator(
    val_tf, val_tf.usage, length=n_input, batch_size=32)

# test generator
test_init_gen = pd.concat([val_tf[-n_input:], test_tf], axis=0)
test_gen = TimeseriesGenerator(
    test_init_gen, test_init_gen.usage, length=n_input, batch_size=len(test_init_gen))
```

### Building the Model

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense

# build the model
lstm_model = keras.models.Sequential()
lstm_model.add(LSTM(300, activation='relu', input_shape=(n_input, n_features), return_sequences=True))
lstm_model.add(LSTM(200, activation='relu'))
lstm_model.add(Dense(100, activation='relu'))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer="adam", loss='mse')
lstm_model.summary()
```

### Fitting the Model

```python
# setup callbacks: model checkpoint and earlystopping

lstm_model_path = os.path.join(BASE_DIR, "lstm_model.h5")

checkpoint_cb = keras.callbacks.ModelCheckpoint(lstm_model_path, save_best_only=True)

early_stopping_cb = keras.callbacks.EarlyStopping(patience=15, 
                                                  restore_best_weights=True)
```

```python
# fit model
lstm_model.fit(train_gen, epochs=100, validation_data=val_gen, 
          callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# plot training and validation loss
pd.DataFrame(lstm_model.history.history).plot(figsize=(12, 4))
plt.xlabel("epoch")
plt.ylabel("loss")
plt.grid(True)
```

### LSTM Model Prediction and Evaluation

```python
# plot training vs. training prediction

# revert back to the best model found during training
lstm_model = keras.models.load_model(lstm_model_path)

# make predictions on the test set
lstm_tr_preds = lstm_model.predict(train_gen)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(train_tf.index, train_tf.usage, label="train")
ax.plot(train_tf.index[:-n_input], lstm_tr_preds, label="LSTM train prediction")
plt.legend()
```

```python
# plot testing vs. test prediction

# make predictions on the test set
lstm_ts_preds = lstm_model.predict(test_gen)

lstm_ts_preds = pd.Series(lstm_ts_preds.flatten(), index=test_tf.index)

fig, ax = plt.subplots(figsize=(12, 4))
test_tf.usage.plot(label="test", color='blue', ax=ax)
lstm_ts_preds.plot(label="LSTM test prediction", 
                        color='orange', ax=ax)

ax.legend(loc="best")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

lstm_model_mae = mean_absolute_error(test_tf.usage, lstm_ts_preds)
lstm_model_mse = mean_squared_error(test_tf.usage, lstm_ts_preds)
lstm_model_rmse = np.sqrt(lstm_model_mse)

print("LSTM MAE: %f"  % lstm_model_mae)
print("LSTM MSE: %f"  % lstm_model_mse)
print("LSTM RMSE: %f"  % lstm_model_rmse)
```

## Deep Learning: GRU Model

### Building the Model

```python
from keras.layers import GRU


gru_model = keras.models.Sequential()
gru_model.add(
    GRU(300, activation="relu", input_shape=(n_input, n_features), 
        return_sequences=True, dropout=0.1, recurrent_dropout=0.1))
gru_model.add(GRU(200, activation='relu', dropout=0.1, recurrent_dropout=0.1))
gru_model.add(Dense(100, activation='relu'))
gru_model.add(Dense(1))
gru_model.compile(optimizer='adam', loss='mse')
gru_model.summary()
```

### Fitting the Model

```python
# setup callbacks: model checkpoint and earlystopping

gru_model_path = os.path.join(BASE_DIR, "gru_model.h5")

checkpoint_cb = keras.callbacks.ModelCheckpoint(
    gru_model_path, save_best_only=True)

early_stopping_cb = keras.callbacks.EarlyStopping(
    patience=15, restore_best_weights=True)
```

```python
# fit model
gru_model.fit(train_gen, epochs=100,
               validation_data=val_gen, 
               callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# plot training and validation loss
pd.DataFrame(gru_model.history.history).plot(figsize=(12, 4))
plt.xlabel("epoch")
plt.ylabel("loss")
plt.grid(True)
```

### GRU Model Prediction and Evaluation

```python
# plot training vs. training prediction

# revert back to the best model found during training
gru_model = keras.models.load_model(gru_model_path)

# make predictions on the test set
gru_tr_preds = gru_model.predict(train_gen)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(train_tf.index, train_tf.usage, label="train")
ax.plot(train_tf.index[:-n_input], gru_tr_preds, label="GRU train prediction")
plt.legend()
```

```python
# plot testing vs. test prediction

# make predictions on the test set
gru_ts_preds = gru_model.predict(test_gen)
gru_ts_preds = pd.Series(gru_ts_preds.flatten(), index=test_tf.index)

fig, ax = plt.subplots(figsize=(12, 4))
test_tf.usage.plot(label="test", color='blue', ax=ax)
gru_ts_preds.plot(label="GRU test prediction", 
                        color='orange', ax=ax)
ax.legend(loc="best")
```

```python
gru_model_mae = mean_absolute_error(test_tf.usage, gru_ts_preds)
gru_model_mse = mean_squared_error(test_tf.usage, gru_ts_preds)
gru_model_rmse = np.sqrt(gru_model_mse)
print("GRU MAE: %f"  % gru_model_mae)
print("GRU MSE: %f"  % gru_model_mse)
print("GRU RMSE: %f"  % gru_model_rmse)
```

## Deep Learning: Bidirectional RNN Modeling

### Building the Model

```python
from keras.layers import Bidirectional

bi_model = keras.models.Sequential()
bi_model.add(Bidirectional(
    LSTM(300, activation="relu", return_sequences=True), input_shape=(n_input, n_features)))
bi_model.add(Bidirectional(
    GRU(200, activation="relu", dropout=0.2, recurrent_dropout=0.2)))
bi_model.add(Dense(100, activation='relu'))
bi_model.add(Dense(1))
bi_model.compile(optimizer='adam', loss='mse')
bi_model.summary()
```

### Fitting the Model

```python
# setup callbacks: model checkpoint and earlystopping

bi_model_path = os.path.join(BASE_DIR, "bi_model.h5")

checkpoint_cb = keras.callbacks.ModelCheckpoint(
    bi_model_path, save_best_only=True)

early_stopping_cb = keras.callbacks.EarlyStopping(
    patience=15, restore_best_weights=True)
```

```python
# fit model
bi_model.fit(train_gen, epochs=100, 
               validation_data=val_gen,
               callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# plot training and validation loss
pd.DataFrame(bi_model.history.history).plot(figsize=(12, 4))
plt.xlabel("epoch")
plt.ylabel("loss")
plt.grid(True)
```

### Bidirectional RNN Model Prediction and Evaluation

```python
# plot training vs. training prediction

# revert back to the best model found during training
bi_model = keras.models.load_model(bi_model_path)

# make predictions on the test set
bi_tr_preds = bi_model.predict(train_gen)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(train_tf.index, train_tf.usage, label="train")
ax.plot(train_tf.index[:-n_input], bi_tr_preds, label="Bidirectional RNN train prediction")
plt.legend()
```

```python
# plot testing vs. test prediction

# make predictions on the test set
bi_ts_preds = bi_model.predict(test_gen)
bi_ts_preds = pd.Series(bi_ts_preds.flatten(), index=test_tf.index)

fig, ax = plt.subplots(figsize=(12, 4))
test_tf.usage.plot(label="test", color='blue', ax=ax)
bi_ts_preds.plot(label="Bidirectional RNN test prediction", 
                        color='orange', ax=ax)
ax.legend(loc="best")
```

```python
bi_model_mae = mean_absolute_error(test_tf.usage, bi_ts_preds)
bi_model_mse = mean_squared_error(test_tf.usage, bi_ts_preds)
bi_model_rmse = np.sqrt(bi_model_mse)
print("Bidirectional MAE: %f"  % bi_model_mae)
print("Bidirectional MSE: %f"  % bi_model_mse)
print("Bidirectional RMSE: %f"  % bi_model_rmse)
```
